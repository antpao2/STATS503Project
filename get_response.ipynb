```python
import pandas as pd
import numpy as np
```

```python
# playoff_result variable: (response)
# 1 → Lost in WC
# 2 → Lost in Divisional
# 3 → Lost in Conference Championship
# 4 → Lost in Super Bowl
# 5 → Won Super Bowl
```

```python
# this function returns the playoff results for each team in the playoff in a given year
def set_response(year):
    playoff_results = pd.read_csv("C:/Users/12074/Desktop/STATS503/Project/" + str(year) + "_playoff_results.csv")
    team_results = pd.DataFrame(columns=['Team', 'Playoff_Result'])
    for i in range(0, playoff_results.shape[0]):
        # Note: There are no ties in the data for the years we are using
        if playoff_results.iloc[i]['Week'] == 'WildCard':
            row = [playoff_results.iloc[i]['Loser/tie'], 1]
            team_results.loc[-1] = row
            team_results.index = team_results.index + 1
            team_results = team_results.sort_index()
        elif playoff_results.iloc[i]['Week'] == 'Division':
            row = [playoff_results.iloc[i]['Loser/tie'], 2]
            team_results.loc[-1] = row
            team_results.index = team_results.index + 1
            team_results = team_results.sort_index()
        elif playoff_results.iloc[i]['Week'] == 'ConfChamp':
            row = [playoff_results.iloc[i]['Loser/tie'], 3]
            team_results.loc[-1] = row
            team_results.index = team_results.index + 1
            team_results = team_results.sort_index()
        elif playoff_results.iloc[i]['Week'] == 'SuperBowl':
            row1 = [playoff_results.iloc[i]['Loser/tie'], 4]
            team_results.loc[-1] = row1
            team_results.index = team_results.index + 1
            team_results = team_results.sort_index()
            row2 = [playoff_results.iloc[i]['Winner/tie'], 5]
            team_results.loc[-1] = row2
            team_results.index = team_results.index + 1
            team_results = team_results.sort_index()
    return team_results
```

```python
years = list(range(2010,2023))
# create a dataframe to store playoff results for all years
all_results = pd.DataFrame(columns=['Team', 'Playoff_Result', 'Year'])
for year in years:
    # get the playoff result dataframe
    playoff_res = set_response(year)
    # set the year 
    playoff_res['Year'] = [year]*playoff_res.shape[0]
    # add to master dataframe
    all_results = pd.concat([all_results, playoff_res], ignore_index=True)
```

```python
# write resulting dataframe to csv
all_results.to_csv("all_results.csv")
```
