# Formatting/merging team_offense and team_defense tables:

```python
# need to format some of the tables
# for 'team_offense' and team_defense tables:
def format_to_td(t_name, col_names):
    years = list(range(2010, 2023))
    res_tab = pd.DataFrame(columns=col_names)
    for year in years:
        # read in each table
        tmp = pd.read_csv("C:/Users/12074/Desktop/STATS503/Project/" + str(year) + "_" + t_name + ".csv")
        # get column names
        tmp.columns = tmp.iloc[0].values
        # drop first row with the col names and get rid of first column which is just the index
        tmp = tmp.iloc[1:, 1:]
        # add column for year
        tmp['Year'] = [year]*tmp.shape[0]
        # merge to res_tab
        res_tab = pd.concat([res_tab, tmp], ignore_index=True)
    # rename columns with indicator for offense/defense
    if t_name == "team_offense":
        res_tab.columns = ['O_' + col for col in col_names]
    elif t_name == "team_defense":
        res_tab.columns = ['D_' + col for col in col_names]
    elif t_name == "conversions":
        res_tab.columns = ['O_' + col for col in col_names]
    elif t_name == "conversions_against":
        res_tab.columns = ['D_' + col for col in col_names]
        
    return res_tab
```

```python
t_name_o = "team_offense"
col_names = ['Rk', 'Tm', 'G', 'PF', 'Yds', 'Ply', 'Y/P', 'TO', 'FL', '1stD', 'Cmp',
       'Att', 'Yds', 'TD', 'Int', 'NY/A', '1stD', 'Att', 'Yds', 'TD', 'Y/A',
       '1stD', 'Pen', 'Yds', '1stPy', 'Sc%', 'TO%', 'EXP', 'Year']
df_o = format_to_td(t_name_o, col_names)
```

```python
t_name_d = "team_defense"
col_names = ['Rk', 'Tm', 'G', 'PA', 'Yds', 'Ply', 'Y/P', 'TO', 'FL', '1stD', 'Cmp',
       'Att', 'Yds', 'TD', 'Int', 'NY/A', '1stD', 'Att', 'Yds', 'TD', 'Y/A',
       '1stD', 'Pen', 'Yds', '1stPy', 'Sc%', 'TO%', 'EXP', 'Year']
df_d = format_to_td(t_name_d, col_names)
df_d.to_csv("team_defense_merged.csv")
```

# Formatting/merging afc_standings/nfc_standings tables:

```python
def format_afc_nfc(t_name, col_names):
    years = list(range(2010, 2023))
    res_tab = pd.DataFrame(columns=col_names)
    for year in years:
        tmp = pd.read_csv("C:/Users/12074/Desktop/STATS503/Project/" + str(year) + "_" + t_name + ".csv")
        # get rid of every 5th row with the division name
        tmp.drop(tmp.index[::5], axis=0, inplace=True)
        # get rid of unneeded columns
        tmp = tmp[["Tm", "W", "L", "PF", "PA", "PD", "MoV", "SoS"]]
        # add column for year
        tmp['Year'] = [year]*tmp.shape[0]
        # merge to res_tab
        res_tab = pd.concat([res_tab, tmp], ignore_index=True)
    # write to csv
    res_tab.to_csv(t_name + "_merged.csv")
        
    return res_tab
```

```python
t_name_afc = "afc_standings"
col_names = ["Tm", "W", "L", "PF", "PA", "PD", "MoV", "SoS", "Year"]
afc_merged = format_afc_nfc(t_name_afc, col_names)
afc_merged
```

```python
# repeat for nfc
t_name_nfc = "nfc_standings"
col_names = ["Tm", "W", "L", "PF", "PA", "PD", "MoV", "SoS", "Year"]
nfc_merged = format_afc_nfc(t_name_nfc, col_names)
nfc_merged
```
