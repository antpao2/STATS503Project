```python
import pandas as pd
import numpy as np
```

```python
# tables needed:
# afc_standings_merged
# nfc_standings_merged
# all_results
# conversions_merged
# team_defense_merged
# team_offense_merged
# conversions_against_merged
```

```python
# get playoff teams
playoff_teams = pd.read_csv("all_results.csv")
# rename team column
playoff_teams.rename(columns = {'Team':'Tm'}, inplace = True)
```

```python
playoff_teams
```

```python
# load all other tables
a_s = pd.read_csv("afc_standings_merged.csv")
n_s = pd.read_csv("nfc_standings_merged.csv")
c = pd.read_csv("conversions_merged.csv")
t_o = pd.read_csv("team_offense_merged.csv")
t_d = pd.read_csv("team_defense_merged.csv")
c_a = pd.read_csv("conversions_against_merged.csv")
```

```python
# rename team column
c.rename(columns = {'O_Tm':'Tm', 'O_Year':'Year' }, inplace = True)
c_a.rename(columns = {'D_Tm':'Tm', 'D_Year':'Year' }, inplace = True)
```

```python
t_d.rename(columns = {'D_Tm':'Tm', 'D_Year':'Year'}, inplace = True)
```

```python
# need to row bind nfc and afc standings tables
s = pd.concat([a_s, n_s], ignore_index=True)
```

```python
# get rid of unnecessary first column for all tables
s = s.iloc[:,1:]
playoff_teams = playoff_teams.iloc[:,1:]
c = c.iloc[:,1:]
t_o = t_o.iloc[:,1:]
t_d = t_d.iloc[:,1:]
c_a = c_a.iloc[:,1:]
```

```python
# need to remove special characters (*) from the 'Tm' columns
s['Tm'] = s['Tm'].str.replace('([^\w/ ])', '')
```

```python
# left join with each table
dfs = [playoff_teams, s, c, c_a, t_o, t_d]
import functools as ft
df_final = ft.reduce(lambda left, right: pd.merge(left, right, on=['Tm', 'Year']), dfs)
```

```python
df_final.to_csv("final_table.csv")
```
